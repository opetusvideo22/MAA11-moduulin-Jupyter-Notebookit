## While-toistorakenne eli silmukka

`while`-silmukka toistaa **sisennetty√§** koodia niin kauan kuin annettu ehto on tosi. Se tarkistaa ehdon **ennen** jokaista toistoa.

Esimerkki:
```python
while ehto:
    # koodi jota toistetetaan kunnes ehto ei ole en√§√§ tosi
```

Silmukka p√§√§ttyy, kun ehto ei en√§√§ ole tosi. Ehdon muotoilussa kannattaa olla varovainen. Mik√§li ehto on aina tosi, silmukka ei p√§√§ty koskaan.

Lue alla olevan ohjelman koodi ja testaile sen toimintaa.

````python
import math

print("T√§m√§ ohjelma laskee lukujen neli√∂juuria. Anna negatiivinen luku lopettaaksesi.")
luku = float(input("Anna luku: "))

while luku >= 0:
    juuri = math.sqrt(luku)
    print("Luvun",luku," neli√∂juuri on",juuri)
    luku = float(input("Anna uusi luku: "))

print("Ohjelma p√§√§ttyi, koska annoit negatiivisen luvun.")
````

### Lis√§esimerkki ja `random`-moduuli

Alla olevassa esimerkiss√§ on hy√∂dynnetty Pythonin satunnaislukuja ja muita satunnaisuuksia tuottavaa `random`-moduulia. Osoitteessa cheat.abitti.fi olevasta oppaasta l√∂yd√§t lis√§tietoja kyseisen moduulin k√§yt√∂st√§.

Lue ja aja alla oleva esimerkkikoodi.

````python
import random

# Arvotaan satunnainen luku 1-10 v√§lilt√§
oikea = random.randint(1, 10)

# Kysyt√§√§n k√§ytt√§j√§√§ arvaamaan lukua
arvaus = int(input("Arvaa luku v√§lilt√§ 1‚Äì10: "))

# Luodaan toistorakenne, joka kysyy lukua kunnes se menee oikein. !=-merkki tarkoittaa erisuurta.
while arvaus != oikea:
    print("V√§√§rin, yrit√§ uudelleen.")
    arvaus = int(input("Arvaa luku v√§lilt√§ 1‚Äì10: "))

print("Oikein! Hyv√§ arvaus.")
````

### Sijoitusoperaattorit

Usein toistorakenteissa eli silmukoissa (eli loopeissa (luupeissa)) lis√§t√§√§n muuttujan arvoon joka kierroksella samaa lukua. T√§ss√§ ja vastaavissa operaatioissa voidaan oikaista **sijoitusoperaattoreiden** avulla.

Esim. `+=` on lyhennysmuoto, joka kasvattaa muuttujan arvoa. Koodissa `x += 1` tarkoittaisi `x = x + 1`.

Alla olevassa taulukossa on listattu tyypillisimm√§t sijoitusoperaattorit.

| Operaattori | Esimerkki   | Merkitys        |
|-------------|-------------|-----------------|
| `+=`        | `luku += 1` | `luku = luku + 1` |
| `-=`        | `luku -= 2` | `luku = luku - 2` |
| `*=`        | `luku *= 3` | `luku = luku * 3` |
| `/=`        | `luku /= 4` | `luku = luku / 4` |
| `%=`        | `luku %= 5` | `luku = luku % 5` |

Mit√§ alla oleva koodi tekee? Lue ja aja.

````python
laskuri = 1

while laskuri <= 50:
    print(laskuri)
    laskuri += 1
````

Miksi alla olevaa koodia ei kannata ajaa?

````python
kakkua_jaljella = 1

while kakkua_jaljella > 0:
    print("Otetaan j√§ljell√§ olevasta kakusta puolet")
    kakkua_jaljella /= 2
````
