## if-elif-else-rakenne

`if–elif–else` on valintarakenne, jossa ohjelma valitsee täsmälleen yhden haaran (tai ei yhtään, jos else puuttuu ja mikään ehto ei täyty). Ehdot arvioidaan **ylhäältä alas**, ja heti kun jokin ehto on tosi, sen osio suoritetaan ja loput ohitetaan.

Rakenne:
```python
if ehto_1:
    # suoritetaan jos ehto_1 on tosi
elif ehto_2:
    # suoritetaan jos ehto_1 oli epätosi ja ehto_2 on tosi
elif ehto_3:
    # ...
else:
    # suoritetaan jos mikään yllä olevista ei ollut tosi
```

Rakenteessa oleva komento `elif` on lyhenne sanoista *else if*.

Alla on esitetty yksinkertainen esimerkki rakenteesta käytössä. Aja koodi ja kokeile muuttaa muuttujan arvoa saadaksesi kaikki eri tulosteet.

````python
arvo = 42

if arvo < 0:
    print("Negatiivinen")
elif arvo == 0:
    print("Nolla")
else:
    print("Positiivinen")
````

`if-elif-else`-rakenteessa on tärkeää asettaa ehdot oikeaan järjestykseen. Muista, että koodi tutkii ja suorittaa ehtojen sisältämän komennot järjestyksessä ylhäältä alas. Alla on kaksi koodia, jotka antavat kokeesta saadun pistemäärän perusteella arvosanan. Molemmat koodit näyttävät pikaisesti katseltuna järkeviltä, mutta vain toinen toimii. **Päättele kumpi toimii lukemalla koodia ja testaa vastauksesi ajamalla koodit.**

````python
# Koodi 1

pisteet = int(input("Anna pistemäärä (0–100): "))

if pisteet >= 90:
    arvosana = "5 (erinomainen)"
elif pisteet >= 75:
    arvosana = "4 (kiitettävä)"
elif pisteet >= 60:
    arvosana = "3 (hyvä)"
elif pisteet >= 45:
    arvosana = "2 (tyydyttävä)"
elif pisteet >= 30:
    arvosana = "1 (välttävä)"
else:
    arvosana = "0 (hylätty)"

print("Arvosana:", arvosana)
````

````python
# Koodi 2

pisteet = int(input("Anna pistemäärä (0–100): "))

if pisteet >= 30:
    arvosana = "1 (välttävä)"
elif pisteet >= 45:
    arvosana = "2 (tyydyttävä)"
elif pisteet >= 60:
    arvosana = "3 (hyvä)"
elif pisteet >= 75:
    arvosana = "4 (kiitettävä)"
elif pisteet >= 90:
    arvosana = "5 (erinomainen)"
else:
    arvosana = "0 (hylätty)"

print("Arvosana:", arvosana)
````
