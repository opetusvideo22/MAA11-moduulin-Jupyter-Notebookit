## 1. Tulostaminen ja kommentit

Ohjelmoinnissa on usein tärkeää **näyttää käyttäjälle tietoa**. Muuten ohjelma vain toimii taustalla tehden jotain joka jää arvoitukseksi. Tätä varten Pythonissa on komento `print()`. Pythonin komennot ovat usein tätä muotoa, joka voi olla tuttu esim. TI-Nspire-laskimesta (esim. solve()-komento).

Aja alla oleva yhden rivin koodi. Se onnistuu klikkaamalla koodisolu aktiiviseksi ja painamalla `Shift-Enter` tai klikkaamalla yläpalkin play-painiketta.


```python
print("Hei maailma!")
```

Huomaa, että edellisessä koodissa teksti kirjoitetaan lainausmerkkien `" "` sisään.

`print()` osaa tulostaa myös lukuja tai laskujen tuloksia. Tällöin lainausmerkkejä ei tarvita. Kokeile alla olevaa koodia ja muokkaa koodissa olevia lukuja, jotta saat erilaisen vastauksen laskulle. Koodi täytyy ajaa uudelleen muokkauksen jälkeen.

```python
print(1 + 1)
```

`print`-komennon sisään voi halutessaan yhdistää tekstiä ja laskuja. Kokeile alla olevaa esimerkkiä.

```python
print("Resepti sanoo, että Niilon tulee syödä kaksi kapselia kolme kertaa päivässä. Tällöin Niilo syö päivässä yhteensä", 2*3, "kapselia.")
```

### Virheet
Koodia kirjoittaessa tulee helposti erilaisia virheitä eli **bugeja**. Alla olevassa koodissa on virhe. Aja se ensin sellaisenaan ja yritä tulkita sen antamaa virhe ilmoitusta. Tämän jälkeen muokkaa koodi toimivaksi.

```python
print("Missä virhe?)
```

Virheilmoitukset eivät aina kerro täsmällisesti mikä virhe on tapahtunut ja missä kohtaa, mutta usein niiden tulkitsemisesta on apua. Älä siis jätä niitä lukematta!

### Kommentit
**Kommentit** ovat rivejä koodissa, joita tietokone ei suorita. Niitä käytetään selittämään koodia tai kirjoittamaan muistiinpanoja.

Pythonissa kommentti aloitetaan merkillä `#`. Kokeile ajaa seuraava koodi ja testaa vaikuttaako kommenttien muokkaaminen koodin toimintaan.


```python
# Tämä on kommentti, tietokone ei suorita tätä

print("Hei maailma!")  # Kommentti voi olla myös rivin lopussa
```

Kommentit ovat hyödyllisiä, kun haluat selittää, mitä jokin koodirivi tekee, esimerkiksi. Varsinkin pitkissä koodeissa kommentointi helpottaa koodin tulkitsemista myöhemmin.

Alla on esimerkkinä koodi, joka etsii Eratosthenesin seulan avulla alkulukuja. Koodia ei tarvitse vielä osata tehdä tai ymmärtää (voisi olla tosin hyvä koetehtävä loppukokeeseen!), mutta komentointi tekee siitä jo nyt huomattavasti ymmärrettävämmän (varsinkin, jos muistaa miten seula itsessään toimii). Kokeile ajaa koodi ja katso mitä tapahtuu.

```python
# Määritetään suurin luku tarkasteluväliltä
n = 120

# Tehdään lista tarkasteltavista luvuista n:nnään asti
luvut=[]
for i in range(2,n+1):
    luvut.append(i)

# Määritetään a
a=2

# Poistetaan listasta a:lla jaolliset a:ta isommat luvut
while n**0.5 > a:
    for luku in luvut:
        if luku % a == 0 and luku > a:
            luvut.remove(luku)
    a = a+1

# Tulostetaan lopputulos listana
print(luvut)
```

## 2. Muuttujat

Ohjelman käsittelemä tieto saadaan talteen **muuttujien** avulla. Niihin voidaan tallettaa mm. lukuja ja tekstiä. Muuttujia voidaan myös päivittää kesken koodin saamaan uusi arvo. Muokkaa alla olevaan koodiin omat tietosi ja aja se.

```python
# Tallennetaan tiedot muuttujiin
nimi = "Maija"
ikä = 17

# Tulostetaan tiedot näkyville
print(nimi, "on", ikä, "vuotta vanha.")
```

**Sääntöjä muuttujien nimeämiseen:**
- Nimi ei voi alkaa numerolla.
- Nimi ei saa sisältää välilyöntejä.
- Yleensä käytetään **pieniä kirjaimia** ja tarvittaessa alaviivaa (`koulun_nimi`).

## 3. Laskutoimitukset ja operaattorit

Python toimii myös yksinkertaisena laskimena ja osaa seuraavat perusoperaatiot (lisää laskuoperaatioita opetellaan tekemään myöhemmin):

| Operaattori | Esimerkki | Tulos      | Selitys             |
|-------------|-----------|------------|---------------------|
| `+`         | `5 + 2`   | 7          | yhteenlasku         |
| `-`         | `5 - 2`   | 3          | vähennys            |
| `*`         | `5 * 2`   | 10         | kertolasku          |
| `/`         | `5 / 2`   | 2.5        | jakolasku |
| `//`        | `5 // 2`  | 2          | jakolaskun kokonaisosa |
| `%`         | `5 % 2`   | 1          | jakojäännös         |
| `**`        | `5 ** 2`  | 25         | potenssi            |

Aja alla oleva esimerkkikoodi.

```python
a = 5
b = 2
print(a + b)
print(a ** b)
```

## 4. Tiedon kysyminen käyttäjältä

`input()` kysyy käyttäjältä tietoa, jonka voi tallentaa muuttujaan. Kokeile ajaa alla oleva koodi ja kirjoita nimesi avautuvaan kenttään ja paina sen jälkeen `Enter`.


```python
nimi = input("Anna nimesi: ")
print("Hei", nimi)
```

`input()` on erityisen kätevä ohjelmissa, joissa halutaan käyttäjällä olevan mahdollisuuksia valita asioita ilman, että koodia itsessään täytyy muokata.

**Huom!** `input()`-komento ottaa annetun syötteen vastaan aina merkkijonona (eng. string, lyh. `str`). Jos annat `input`-komennolle lukuja ja yrität laskea niillä, se ei onnistu, koska Python kuvittelee niiden olevan vain kirjainmerkkejä. Lue alla oleva koodi ja kokeile sen toimintaa.


```python
# Kerrotaan käyttäjälle mitä ohjelma tekee
print("Tämä ohjelma laskee antamasi kahden luvun tulon")

# Pyydetään käyttäjältä luvut ja tallennetaan muuttujiin a ja b
a = input("Anna ensimmäinen luku: ")
b = input("Anna toinen luku: ")

# Tulostetaan vastaus
print("Lukujen tulo on:", a*b)
```

Huomaat, että virhekoodi kertoo seuraavaa: _TypeError: can't multiply sequence by non-int of type 'str'_. Opetellaan seuraavaksi muut **muuttujatyypit** ja miten edellisen kohdan ongelma voidaan ratkaista.

## 5. Muuttujatyypit: `int`, `float` ja `str`

Pythonissa on useita **muuttujatyyppejä**, joista tärkeimmät ovat:
- **int** (integer) = kokonaisluku, esim. `5`
- **float** = desimaaliluku, esim. `3.14`
- **str** (string) = teksti, esim. `"Hei"`

Joskus on tarpeen **muuntaa** tietoa tyypistä toiseen. Esimerkiksi kun käyttäjän antama tieto `input()`-komennolla on aina tekstiä (`str`), voi sen muuttaa kokonaisluvuksi komennolla `int()`, mikäli mahdollista.


```python
luku = int(input("Anna kokonaisluku ja lisään siihen 7"))
print(luku+7)
```

Kokeile syöttää edelliseen koodiin luvuksi kokonaisluvun sijaan desimaaliluku (esim. 3.14). Huomaa, että Python käyttää desimaalierottimena pistettä. **Mitä tapahtui?**

Jos haluat muuttaa `input`-komennolle annetun merkkijonon (`str`) desimaaliluvuksi, se tapahtuu `float()`-komennolla. Alla edellisen esimerkin koodi muokattuna toimimaan myös desimaaliluvuilla. Testaa sen toimintaa kokonais- ja desimaaliluvuilla.

```python
luku = float(input("Anna luku ja lisään siihen 7"))
print(luku+7)
```

Tee koodi alla olevaan tyhjään koodisoluun ja aja se.

## 6. Pyöristäminen

Mikäli koodissa esiintyvää lukua halutaan pyöristää, se onnistuu `round()`-komennolla. Komennon sisään merkitään pyöristettävä luku ja pyöristyksen tarkkuus pilkulla erotettuna. Kokeile alla olevaa esimerkkiä ja päättele, miten pyöristyksen tarkkuuden säätäminen toimii.


```python
luku = 6.14159

print(round(luku, 2))
print(round(luku, 0))
print(round(luku, 4))
print(round(luku, -1))
```
