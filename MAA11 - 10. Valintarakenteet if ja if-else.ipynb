# MAA11 – 10. Valintarakenteet `if` ja `if-else`

Tämän tunnin tavoitteena on oppia käyttämään **if**- ja **if-else** -rakenteita Pythonissa. Lisäksi tutustumme vertailuoperaattoreihin, **and**- ja **or**-ehtojen käyttöön ja siihen, miten voidaan käyttää sisäkkäisiä if-lauseita.

## 1. IF-lause
**If-lause** on ohjelmoinnin perusrakenne, jonka avulla voidaan suorittaa tiettyjä käskyjä vain, **jos** jokin ehto on totta. Tämä mahdollistaa ohjelman toiminnan muuttamisen tilanteen mukaan.

Rakenne:
```python
if ehto:
    # Tämä koodi, joka tulee olla sisennettynä, suoritetaan vain, jos ehto on totta
```

Useimmat ohjelmointiympäristöt tekevät sisennyksen (eng. _indentation_) automaattisesti, kun kirjoitat kaksoispisteen `:` ja painat Enteriä. Sisennyksestä voi poistua painamalla Backspace-näppäintä. Älä siis lähde tekemään sisennystä itse välilyönnein, jos sisennys syntyy itsestään. Näin vältät bugeja koodissa.

**Aja alla oleva esimerkki** ja katso, mitä tapahtuu, kun annat syötteeksi positiivisia tai negatiivisia kokonaislukuja.

````python
luku = int(input("Anna kokonaisluku ja kerron, jos se on positiivinen."))

if luku > 0:
    print('Luku on positiivinen.')
````

## 2. Vertailuoperaattorit
Edellisessä esimerkissä käytettiin **vertailuoperaattoria** `>`. If-lauseen ehto määritetään yleensä vertaamalla arvoja keskenään. Pythonissa käytetään seuraavia vertailuoperaattoreita:

- `==` : on yhtä suuri kuin
- `!=` : ei ole yhtä suuri kuin
- `<`  : pienempi kuin
- `>`  : suurempi kuin
- `<=` : pienempi tai yhtä suuri kuin
- `>=` : suurempi tai yhtä suuri kuin

Erityishuomiota on syytä antaa operaattorille `==`, sillä se todellakin tarvitsee kaksi yhtä suuruus -merkkiä.

**Aja seuraava esimerkkikoodi.**

````python
x = 5

print(x < 5)
print(x == 5)
print(x >= 5)
````

Edellisen esimerkin tulosteessa törmäämme uuteen **muuttujatyyppiin**. Viime kerralla opimme jo kolme erilaista: `int`, `float` ja `str`. Tällä kertaa luvassa on `bool`, joka edustaa totuusarvoja. Se voi saada arvokseen siis `True` (suom. tosi) tai `False` (suom. epätosi). Joskus näiden tilalla käytetään lukua 1 ja 0, jonka oletkin jo huomannut tai tulet huomaamaan MAA11-moduulin aikana. Tähän muuttujatyyppiin palaamme uudelleen myöhemmin.

**Harjoitustehtävä**: Kokonaisluku on jaollinen toisella kokonaisluvulla, jos **jakojäännös** on nolla. Kirjoita koodi, joka kysyy käyttäjältä positiivista kokonaislukua ja ilmoittaa, tulosteella _Lukusi on jaollinen luvulla 5._, mikäli se tosiaan on jaollinen luvulla 5. Mikäli et muista miten jakojäännös laskettiin Pythonissa, käy lunttaamassa se osoitteesta [cheat.abitti.fi](https://cheat.abitti.fi/). Tee koodi alla olevaan tyhjään koodisoluun ja aja se.

## 3. IF-ELSE-lause
Joskus haluamme ohjelman tekevän jonkin muun toiminnon silloin, kun ehto ei ole totta. Tällöin käytetään **else**-osaa. If-else-rakenne tarkoittaa: *Jos ehto on totta, tee tämä. Muussa tapauksessa tee jotain muuta.*

Rakenne:
```python
if ehto:
    # Tämä koodi, joka on sisennettynä, suoritetaan, jos ehto on totta
else:
    # Tämä koodi, joka on sisennettynä, suoritetaan, jos if-lauseen ehto ei ole totta
```

**Aja alla oleva esimerkkikoodi**, joka on muokattu versio aikaisemmasta koodista.


````python
luku = int(input("Anna kokonaisluku ja kerron, jos se on positiivinen."))

if luku > 0:
    print("Luku on positiivinen.")
else:
    print("Luku ei ole positiivinen.")
````

Yhdistämällä `if`-komennon kaveriksi `else`, saamme aikaan järkevämmin toimivaa koodia.

**Harjoitustehtävä**: Olet saanut kesätöitä kaupan kassalta. Tee itsellesi ohjelma, jolla voit helpottaa asiakkaiden iän tarkistamista, mikäli he ovat ostamassa alaikäisiltä kiellettyjä tuotteita. **Iän tarkastamiseksi riittää tässä tehtävässä tarkastella vain syntymävuotta** (parannellaan tätä ohjelmaa hetken päästä). Voit toteuttaa ohjelman haluamallasi tavalla, mutta sen tulee antaa joko tuloste _Älä myy kaljaa_ tai _Myy kaljaa_ riippuen asiakkaan syntymävuodesta.

Tähän asti tekemämme koodit ovat olleet melko lyhyitä, joten on hyvä tiedostaa, että usein näin ei ole. `if`- ja `if-else`-rakenteet ovat usein vain osa koodia ja koodi voi jatkua niiden jälkeenkin. **Python käy läpi koodia ylhäältä alaspäin rivi kerrallaan** paitsi esim. silmukoiden kohdalla, joihin palaamme myöhemmin.

**Lue** alla oleva esimerkkikoodi ja **hahmota** mielessäsi, miten se tulee toimimaan. Seuraavaksi **aja** alla koodi ja katso miten käy.

````python
luku = int(input("Anna kokonaisluku ja kerron, jos se on positiivinen."))

if luku > 0:
    print("Luku on positiivinen.")
else:
    print("Luku ei ole positiivinen.")

print("Kokeillaan seuraavaksi ihan huvikseen, onko antamasi luku jaollinen kympillä.")

if luku % 10 == 0:
    print("No onhan se :D")
else:
    print("No eipä näytä olevan")

print("Eiköhän tämä riitä ja esimerkin idea tullut selville :)")
````

## 5. AND ja OR -ehtojen yhdistäminen
Joskus haluamme tarkistaa useita ehtoja yhtä aikaa. **and** tarkoittaa, että molempien ehtojen täytyy olla totta. **or** tarkoittaa, että riittää, että toinen ehto on totta. Kokeile alla olevat esimerkit läpi ja tutki, miten ehdot toimivat.

````python
ika = 20
kansalaisuus = "Suomi"

if ika >= 18 and kansalaisuus == "Suomi":
    print("Voit äänestää Suomessa!")
````

````python
paiva = input("Mikä päivä tänään on? Valitse jokin vaihtoehdoista: ma, ti, ke, to, pe, la, su:   ")

if paiva == "la" or paiva == "su":
    print("On viikonloppu!")
else:
    print("On arkipäivä.")
````

## 6. Sisäkkäinen IF
Joskus tarvitaan **if-lauseen sisälle toinen if-lause**, kun halutaan tarkistaa useampi ehto peräkkäin. Tätä kutsutaan sisäkkäiseksi rakenteeksi. **Tutustu** alla olevaan esimerkkikoodiin ja **aja** se useasti siten, että saat aikaan kaikki kolme eri tulostetta.

````python
# Sisäkkäinen if-esimerkki
luku = int(input("Anna kokonaisluku."))

if luku > 0:
    if luku >= 100:
        print('Luku on positiivinen ja suurempi tai yhtä suuri kuin 100')
    else:
        print('Luku on positiivinen, mutta pienempi kuin 100')
else:
    print("Luku on negatiivinen tai nolla.")
````

Seuraavalla tunnilla opiskellaan `if-elif-else`-rakenne, joka helpottaa vastaavanlaisissa tilanteissa koodin tekemistä ja tietokoneen kuormittamista.
